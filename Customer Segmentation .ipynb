```python
x=df.iloc[:,[3,4]].values #:-row
```

```python
x
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
k_values = range(1, 11)

# Fit the k-means model for each k and calculate inertia
inertia = []
for k in k_values:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(x)
    inertia.append(kmeans.inertia_)

# Plot the elbow curve
plt.plot(k_values, inertia, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Sum of Squared Distances (Inertia)')
plt.title('Elbow Method for Optimal k')
plt.show()
```

```python
kmeans =KMeans(n_clusters=5, init='k-means++',random_state=1)
```

```python
y=kmeans.fit_predict(x)
```

```python
y
```

```python
plt.figure(figsize=(5,5))
plt.scatter(x[y==0,0],x[y==0,1],s=50,c='green',label="c1")
plt.scatter(x[y==1,0],x[y==1,1],s=50,c='red',label="c2")
plt.scatter(x[y==2,0],x[y==2,1],s=50,c='yellow',label="c3")
plt.scatter(x[y==3,0],x[y==3,1],s=50,c='cyan',label="c4")
plt.scatter(x[y==4,0],x[y==4,1],s=50,c='blue',label="c5")
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,c='violet',label='Centroids')
plt.title("customer groups")
plt.xlabel("annual income")
plt.ylabel("spending score")
plt.show()
```
